```python
df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
pd.set_option("display.max.columns", None)

df.head(10)
```

```python
%matplotlib
```

```python
%matplotlib inline
```

```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"], kind="line")
```

```python
df.plot.line(x="Rank", y=["P25th", "Median", "P75th"])
```

```python
median_column = df["Median"]
type(median_column)
```

```python
median_column.plot(kind="hist")
```

```python
top_5 = df.sort_values(by="Median", ascending=False).head()
top_5
```

```python
top_5.plot(x="Major", y="Median", kind="bar")
```

```python
top_5.plot(x="Major", y="Median", kind="bar", rot=5, fontsize=5)
```

```python
top_medians = df[df["Median"] > 60_000].sort_values("Median")
top_medians.plot(x="Major", y=["P25th", "Median", "P75th"], kind="bar", rot=5, fontsize=5)
```

```python
df.plot.line(x="Rank", y=["College_jobs", "Non_college_jobs"])
```

```python
df.plot.scatter(x="Median", y="Unemployment_rate")
```

```python
df.corr(x="Median", y="Unemployment_rate")
```
